```python
import stlabutils
```

```python
mydata = stlabutils.readdata.readdat(
            '../examples/data/M59_2017_06_26_16.58.40_RF_vs_power_m60dbmatt_2amp_ref_sample.dat')
```

```python
list(mydata[0]) == ['Frequency (Hz)',
 'S21re ()',
 'S21im ()',
 'S21dB (dB)',
 'S21Ph (rad)',
 'Power (dBm)',
 'Temperature (K)',
 'Power (K)']
```

```python
mydata[0].shape
```

```python
datavna = mydata[-1]
```

```python
#************************************************
fitwidth = 5.  #Width of trace to fit in units of detected peak width
trimwidth = 3.  #Width to trim around detected peak for background fit in units of detected peak width
ftype = 'A'  # A and -A are reflection models while B and -B are side coupled models
doplots = False  #Show debugging plots
margin = 51  #smoothing margin for peak detection
#************************************************
```

```python
#*************************** DO FIT AND RETURN RESULT *********************************************************
z = datavna['S21re ()'] + 1j * datavna['S21im ()']  #Convert S parameter data from Re,Im to complex array
x = datavna['Frequency (Hz)']
fitres, _, _, _ = stlabutils.S11fit(x,z)#,ftype=ftype,doplots=doplots,trimwidth=trimwidth,fitwidth=fitwidth,margin=margin)  #Do fit with some given parameters.  More options available.
fnew = fitres['f0'].value
```

```python
fitres
```

```python
zfit = stlabutils.S11func(datavna['Frequency (Hz)'],fitres,ftype='A')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(datavna['Frequency (Hz)'],datavna['S21dB (dB)'])
plt.plot(datavna['Frequency (Hz)'],20*np.log10(np.abs(zfit)))
plt.show()
plt.close()
```

```python
freqs = datavna['Frequency (Hz)']
thedata = datavna['S21dB (dB)']
thefit = 20*np.log10(np.abs(zfit))

plt.plot(freqs,thedata-thefit)
plt.show()
plt.close()
```

```python
np.sqrt(np.sum((thedata-thefit)**2))
```

```python
ress = stlabutils.utils.S11fit.S11residual(fitres,datavna['Frequency (Hz)'],datavna['S21dB (dB)'])
```

```python
reress=ress[::2]
imress=ress[1::2]
theress=np.abs(reress+1j*imress)
```

```python
plt.plot(datavna['Frequency (Hz)'],theress)
plt.show()
plt.close()
```

```python
plt.plot(datavna['Frequency (Hz)'],imress)
plt.show()
plt.close()
```

```python
plt.plot(datavna['Frequency (Hz)'],reress)
plt.show()
plt.close()
```
